```python
import serial
import time
```

```python
ser = serial.Serial(
    port = "COM13",
    baudrate = 9600,
    bytesize = serial.EIGHTBITS,
    parity = serial.PARITY_NONE,
    stopbits = serial.STOPBITS_ONE,
    timeout = 1.0
)
```

```python
with open("../../EBEAM-Dashboard-Logs/beamEnergy.txt", "w", encoding="utf-8") as f:
    while True:
        line = ser.readline()
        decoded_line = line.decode(errors='ignore').strip()
        f.write(decoded_line + "\n")
```
